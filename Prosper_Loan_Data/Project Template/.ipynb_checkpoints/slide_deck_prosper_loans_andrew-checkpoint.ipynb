## Investigation Overview

> As an active investor on the Prosper Platform since 2012, I have developed several filters to leverage and build on Prosper's proprietary loan grading system, in order to achieve greater overall rates of return. The overarching goal of this project is to investigate the utility of my filters in relation to overall loan outcome.

## Dataset Overview

> The original dataset contains nearly 114,000 records pertaining to loans on the Prosper Platform. Each record represents a loan on the Prosper marketplace, and contains various information regarding the loan and the loan applicant. For example, for each loan the dataset contains information regarding the amount, interest rate, term, and status (Current, Defaulted, Cancelled, etc.). For my scope of investigation, I will be narrowing the set of records to only include loans finalized notes with a ProsperScore (i.e. notes with a non-null 'ClosedDate' that originated after 2009). The application of these filters narrows down the set of records available for analysis to 26,005.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  matplotlib.ticker import PercentFormatter
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('../prosperLoanData.csv')
dfClean =  df.dropna(subset=["ClosedDate"])
dfClean =  dfClean.dropna(subset=["ProsperScore"])
dfClean.shape
```

```python
# Creating bands on Bank Card Utilization
bankCardUtilization_bins = [0,0.22,0.54,0.81,2.5]
bankCardUtilization_bins_labels = ['Quartile 1','Quartile 2','Quartile 3','Quartile 4']
dfClean['BankCardUtilizationBin'] = pd.cut(dfClean['BankcardUtilization'], bins = bankCardUtilization_bins, labels = bankCardUtilization_bins_labels)

# Creating bands based on debt to income
debtToIncome_bins = [0,0.13,0.20,0.30,10.01]
debtToIncome_bins_labels = ['Quartile 1','Quartile 2','Quartile 3','Quartile 4']
#['Low','Moderate','High','Very High']
dfClean['DebtToIncomeRatioBin'] = pd.cut(dfClean['DebtToIncomeRatio'], bins = debtToIncome_bins, labels = debtToIncome_bins_labels)


# Creating bands on open credit lines
ProsperScore_bins = [1,6,11]
ProsperScore_labels =  ['Bottom Half','Top Half']
dfClean['ProsperScoreBin'] = pd.cut(dfClean['ProsperScore'],bins =ProsperScore_bins ,labels = ProsperScore_labels)
```

## Prosper Loan Status

> The overall universe of records of analysis totals roughly 26,000 records, representing thousands of closed loans! For the universe of closed loans reviewed, roughly 76% of loans closed successfully according to the terms of the note, 21% of loans closed with a status of charged off, and 4% defaulted.

```python
plt.figure(figsize = [16, 5])

dfBaseline = dfClean.groupby('LoanStatus')['LoanNumber'].count()
dfBaseline = dfBaseline.divide(dfBaseline.sum())*100
ax = dfBaseline.plot(kind='bar');
plt.title('Prosper Loan Outcome (Closed Loans Only)', size=18)
plt.ylabel('Proportion of Loans (%)', fontsize=16)
plt.xlabel('Loan Status', fontsize=16)

for p in ax.patches:
    value = round(p.get_height(),1)
    #ax.text(str(value), (p.get_x(), p.get_height()))
    ax.text( p.get_x() + .2,p.get_height() - 3, str(value) + '%', color='white', fontweight='bold')
```

## Prosper Loan Status by Prosper Score

> In order to provide investors with an easy way to identify loan riskiness, Prosper has developed a loan scoring system that provides a single value to easily assess the relative risk of one loan versus another. All things equal, higher Prosper Scores correspond to lower rates of default or charge off, while lower scores correlate with higher levels of default and/or charge offs as shown in the below chart.

```python
def get_bar_height(height):
    if height <= 10.0:
        return 100.0 - height
    elif height >= 50.0:
        return height
    else:
        return height - 5.0;
```

```python
dfProsperScoreHalf = dfClean.groupby(['ProsperScoreBin','LoanStatus'])['LoanNumber'].count().unstack()
dfProsperScoreHalf = dfProsperScoreHalf.T.div(dfProsperScoreHalf.T.sum()).T*100
ax = dfProsperScoreHalf.plot(kind='bar', stacked = True);
# Set general font size
plt.rcParams['font.size'] = '10'
plt.title('Loan Status by Prosper Score Halfs', size=18)
plt.ylabel('Proportion of Loans (%)', fontsize=16)
plt.xlabel('Debt to Income Quartile', fontsize=16)

for p in ax.patches:
    value = round(p.get_height(),1)
    get_bar_height(p.get_height())
    bar_height = get_bar_height(p.get_height())
    #ax.text(str(value), (p.get_x(), p.get_height()))
    ax.text( p.get_x() + .2,bar_height, str(value) + '%', color='black', fontweight='bold');

```

## Can We Do Better?
>As Prosper loans are unsecured financial assets whose max return is fixed and max loss is 100%, one of the primary ways an investor can increase returns is to minimize the proportion of loans that enter default or charge off. As previously shown, Prosper Score is the single largest predictor of terminal loan state, however, can other variables add incremental performance? To investigate this topic, we divided the universe of closed loans into two main groups by Prosper Score: Bottom Half (Prosper Score of 6 or less) and Top Half (Prosper Score of 7 or greater). Additionally, for variables of interest, the universe of closed loans are segmented into quartiles with the first quartile representing the bottom 25% of the distribution for a given variable, and subsequent quartiles holding the next 25% of the variables observed distribution.

Note quartile segments were based on quartile ranges resulting from the Pandas dataframe.describe() function.

## Alternative Variables: Bank Card Utilization
> Bank card utilization represents the ratio of an individual's outstanding credit card balance to their total card limits. Given that many individuals use Prosper Loans to refinance high interest rate credit card debt into lower rates, it's not obvious that high card utilization rates are correlated with higher rates of default and charge off in loans after controlling for overall Prosper Scores.

Bank Card Utilization Quartile Ranges
<ul>
    <li>Q1: 0.0% to 22%</li>
    <li>Q2: 22% to 54%</li>
    <li>Q3: 54% to 81%</li>
    <li>Q4: Greater than 81%</li>
</ul>

The following chart shows mixed results for higher Bank Card Utilization rates and loan status. For the bottom half of Prosper Scores, the relationship is flat; however, for the top half of Prosper Scores, lower bank card utilization rates appear to result in better loan outcomes.

```python
# create two data frames for the higher and lower Prosper score bins
dfBottomHalf = dfClean.query("ProsperScoreBin == 'Bottom Half'").groupby(['BankCardUtilizationBin','LoanStatus'])['LoanNumber'].count().unstack()
dfTopHalf = dfClean.query("ProsperScoreBin == 'Top Half'").groupby(['BankCardUtilizationBin','LoanStatus'])['LoanNumber'].count().unstack()

# get loan counts by loan status across bank card utilization quartiles
dfBottomHalf = dfBottomHalf.T.div(dfBottomHalf.T.sum()).T*100 
# divide group counts by totals to scale to 100%
dfTopHalf = dfTopHalf.T.div(dfTopHalf.T.sum()).T*100


# Chargedoff, Completed,Defaulted
fig, ax = plt.subplots(figsize=[30,12])
width = 0.4
# Set general font size
plt.rcParams['font.size'] = '18'

# create dictionary for legend
colors = {'Defaulted':'g','Completed':'orange','Chargedoff':'b'}
# create list from keys of color dict
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color = colors[label]) for label in labels]

plt.subplot(1,2,1)
p1_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Chargedoff'], width, color='b')
p2_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Completed'], width, bottom=dfBottomHalf['Chargedoff'], color='orange')
p3_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Defaulted'], width,
             bottom=[sum(x) for x in zip(dfBottomHalf['Chargedoff'],dfBottomHalf['Completed'])], color='g')
# set title
plt.title('Bank Card Utilization vs Loan Status - Prosper Low Score', size = 30)
plt.ylabel('Proportion of Loans (%)', fontsize=20)
plt.xlabel('Bank Card Utilization Quartile', fontsize=20);

# set legend
plt.legend(handles, labels);

plt.subplot(1,2,2)
p1_top = plt.bar(dfTopHalf.index, dfTopHalf['Chargedoff'], width, color='b')
p2_top = plt.bar(dfTopHalf.index, dfTopHalf['Completed'], width, bottom=dfTopHalf['Chargedoff'], color='orange')
p3_top = plt.bar(dfTopHalf.index, dfTopHalf['Defaulted'], width,
             bottom=[sum(x) for x in zip(dfTopHalf['Chargedoff'],dfTopHalf['Completed'])], color='g')
# set title
plt.title('Bank Card Utilization vs Loan Status - Prosper High Score', size = 30);
# set labels
plt.ylabel('Proportion of Loans (%)', fontsize=20)
plt.xlabel('Bank Card Utilization Quartile', fontsize=20);

# set legend
plt.legend(handles, labels);
```

Debt to Income Quartile Ranges
<ul>

    <li>Q2: 13% to 20%</li>
    <li>Q3: 20% to 30%</li>

</ul>
The following chart shows positive results for Debt to Income Ratios and loan status. For the the bottom and top half of Prosper Scores, lower debt to income ratios are associated with better loan outcomes.

```python
# create two data frames for the higher and lower Prosper score bins
dfBottomHalf = dfClean.query("ProsperScoreBin == 'Bottom Half'").groupby(['DebtToIncomeRatioBin','LoanStatus'])['LoanNumber'].count().unstack()
dfTopHalf = dfClean.query("ProsperScoreBin == 'Top Half'").groupby(['DebtToIncomeRatioBin','LoanStatus'])['LoanNumber'].count().unstack()

# get loan counts by debt to income quartiles
dfBottomHalf = dfBottomHalf.T.div(dfBottomHalf.T.sum()).T*100
# divide group counts by totals to scale to 100%
dfTopHalf = dfTopHalf.T.div(dfTopHalf.T.sum()).T*100


# Chargedoff, Completed,Defaulted
fig, ax = plt.subplots(figsize=[30,12])
width = 0.4
# Set general font size
plt.rcParams['font.size'] = '18'

# create dictionary for legend
colors = {'Defaulted':'g','Completed':'orange','Chargedoff':'b'}
# create list from keys of color dict
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color = colors[label]) for label in labels]

plt.subplot(1,2,1)
p1_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Chargedoff'], width, color='b')
p2_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Completed'], width, bottom=dfBottomHalf['Chargedoff'], color='orange')
p3_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Defaulted'], width,
             bottom=[sum(x) for x in zip(dfBottomHalf['Chargedoff'],dfBottomHalf['Completed'])], color='g')
plt.title('Debt to Income Ratio vs Loan Status - Prosper Low Score', size = 30)
plt.ylabel('Proportion of Loans (%)', fontsize=20);
plt.xlabel('Debt to Income Quartile', fontsize=20);

# set legend
plt.legend(handles, labels);

plt.subplot(1,2,2)
p1_top = plt.bar(dfTopHalf.index, dfTopHalf['Chargedoff'], width, color='b')
p2_top = plt.bar(dfTopHalf.index, dfTopHalf['Completed'], width, bottom=dfTopHalf['Chargedoff'], color='orange')
p3_top = plt.bar(dfTopHalf.index, dfTopHalf['Defaulted'], width,
             bottom=[sum(x) for x in zip(dfTopHalf['Chargedoff'],dfTopHalf['Completed'])], color='g')
# set title
plt.title('Debt to Income Ratio vs Loan Status - Prosper High Score', size = 30);
# set labels
plt.ylabel('Proportion of Loans (%)', fontsize=20);
plt.xlabel('Debt to Income Quartile', fontsize=20);

# set legend
plt.legend(handles, labels);
```

## Controlling for Prosper Score
> Though the above visuals attempt to control for the effect of the dominant factor--low or high Prosper Score--by segmenting the universe of closed loans into two separate halves, it's possible that the effect of Prosper Score is so dominate that the observed relationships are due simply to relative levels of Prosper Scores. To control for this, we focus on two Prosper Score ranges (5 and 9) of approximate equal size. This will allow us to focus on the impact of higher/lower rates of bank card utilization and debt to income ratio on final loan outcome.

### Bank Card Utilization
> Isolating effect of Prosper Score by focusing on only a single Prosper score value for low and high Prosper Scores.

Bank Card Utilization Quartile Ranges
<ul>
    <li>Q1: 0.0% to 22%</li>
    <li>Q2: 22% to 54%</li>
    <li>Q3: 54% to 81%</li>
    <li>Q4: Greater than 81%</li>
</ul>

```python
# Selecting 5 and 9 since they have roughly the same number of closed loans.
# Create new dataframe with only records having propser score of 5 or 9
dfFocused = dfClean.query('ProsperScore == 5 or ProsperScore == 9')

# Create a df with only scores of 5 and group by bank card utilization quartiles
dfBottomHalf = dfFocused.query("ProsperScore == 5").groupby(['BankCardUtilizationBin','LoanStatus'])['LoanNumber'].count().unstack()
# Create a df with only scores of 9 and group by bank card utilization quartiles
dfTopHalf = dfFocused.query("ProsperScore == 9").groupby(['BankCardUtilizationBin','LoanStatus'])['LoanNumber'].count().unstack()

# Scale to 100% by dividing by seprate totals
dfBottomHalf = dfBottomHalf.T.div(dfBottomHalf.T.sum()).T*100
dfTopHalf = dfTopHalf.T.div(dfTopHalf.T.sum()).T*100


# Chargedoff, Completed,Defaulted
fig, ax = plt.subplots(figsize=[30,12])
width = 0.4
# Set general font size
plt.rcParams['font.size'] = '18'

# create dictionary for legend
colors = {'Defaulted':'g','Completed':'orange','Chargedoff':'b'}
# create list from keys of color dict
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color = colors[label]) for label in labels]

plt.subplot(1,2,1)
# Invididually create each bar for each loan status, they are then plotted ontop of each other for each quartile
p1_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Chargedoff'], width, color='b')
p2_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Completed'], width, bottom=dfBottomHalf['Chargedoff'], color='orange')
p3_bottom = plt.bar(dfBottomHalf.index, dfBottomHalf['Defaulted'], width,
             bottom=[sum(x) for x in zip(dfBottomHalf['Chargedoff'],dfBottomHalf['Completed'])], color='g')
plt.title('Bank Card Utilization vs Loan Status - Prosper Score 5', size = 30)
plt.ylabel('Proportion of Loans (%)', fontsize=20)
plt.xlabel('Bank Card Utilization Quartile', fontsize=20)

# set legend
plt.legend(handles, labels);

plt.subplot(1,2,2)
# Invididually create each bar for each loan status, they are then plotted ontop of each other for each quartile
p1_top = plt.bar(dfTopHalf.index, dfTopHalf['Chargedoff'], width, color='b')
p2_top = plt.bar(dfTopHalf.index, dfTopHalf['Completed'], width, bottom=dfTopHalf['Chargedoff'], color='orange')
p3_top = plt.bar(dfTopHalf.index, dfTopHalf['Defaulted'], width,
             bottom=[sum(x) for x in zip(dfTopHalf['Chargedoff'],dfTopHalf['Completed'])], color='g')
# set title
plt.title('Bank Card Utilization vs Loan Status - Prosper Score 9', size = 30);
# set labels
plt.ylabel('Proportion of Loans (%)', fontsize=20)
plt.xlabel('Bank Card Utilization Quartile', fontsize=20);

# set legend
plt.legend(handles, labels);
```

## The effect of Prosper Score on Alternative Variables
> A valid question centers on the fact that high/low rates of bank card utilization and debt to income ratio are associated with lower/higher Prosper Scores. Effectively, are the mean values for Bank Card Utilization and Debt to Income significantly different between the low and high prosper score group? 

To address this question, the below chart plots the mean and standard deviation for Bank Card Utilization and Debt to Income ratio for the Bottom Half Prosper Score group and Top Half Prosper Score group. From the chart, it's clear that the means are different, however, given the variance of both groups the difference in means does not appear to be significant. With this we can more confidently assert that the observed difference in loan status stems from an underlying relationship between Bank Card Utilization and Debt to Income Ratio on loan status not captured by the Proposer Score.

```python
fig, axes = plt.subplots(1, 2, figsize=(30, 12),sharey=True)
fig.suptitle('Mean Bank Card Utilization and Debt To Income by Prosper Score Group')
plt.rcParams['font.size'] = '20'

# Bank Card Utilizatino
sb.barplot(ax=axes[0], x=dfClean['ProsperScoreBin'], y=dfClean['BankcardUtilization'], ci='sd')
axes[0].set_title('Bank Card Utilization (%)')
axes[0].yaxis.set_major_formatter(PercentFormatter(1))


# Debt to Income
dti = sb.barplot(ax=axes[1], x=dfClean['ProsperScoreBin'], y=dfClean['DebtToIncomeRatio'], ci='sd')
axes[1].set_title('Debt to Income (%)');
axes[1].yaxis.set_major_formatter(PercentFormatter(1));
```

## What's the Take Away?

Based on this analysis it appears that the use of Bank Card Utilization and Debt to Income Ratio could help investors identify loans with a lower probability of ending in default or charge off, even after accounting for Proser Score. Said another way, an investor can reduce the number of loans that end in default or charge off and increase overall returns by investing in loans where the borrower has a bank card utilization rate below 22% and debt to income ratio below 13% (ie the bottom quartile for each variable).
