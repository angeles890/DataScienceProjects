## Data Wrangling
Before further exploration can take place, we'll need to:
<ul>   
    <li>Create buckets for certain varialbes to support buckets for analysis</li>
    <li>Update listingCategory to text value </li>
</ul>
1. Create Bands

```python
# Creating bans on debt to income
debtToIncome_bins = [0,0.13,0.2,0.3,0.5,11]
debtToIncome_bins_labels = ['Very Low','Low','Moderate','High','Very High']
dfClean['DebtToIncomeBin'] = pd.cut(dfClean['DebtToIncomeRatio'], bins = debtToIncome_bins, labels = debtToIncome_bins_labels)

dfClean[['DebtToIncomeRatio','DebtToIncomeBin']].sample(5)
```

```python
# Creating bands on open credit lines
openCreditLines_bins = [0,5,8,11,17,48]
openCreditLines_labels = ['Very Low','Low','Moderate','High','Very High']
dfClean['CreditLinesBin'] = pd.cut(dfClean['OpenCreditLines'],bins =openCreditLines_bins ,labels = openCreditLines_labels)
dfClean[['OpenCreditLines','CreditLinesBin']].sample(5)
```

```python
# Create bands on loan amount
loanAmount_bins = [1000,3000,4500,8000,35000]
loanAmount_bins_labels = ['Quartile 1','Quartile 2','Quartile 3','Quartile 4']
dfClean['LoanOriginalAmountBin'] = pd.cut(dfClean['LoanOriginalAmount'],bins =loanAmount_bins ,labels = loanAmount_bins_labels)
dfClean[['LoanOriginalAmount','LoanOriginalAmountBin']].sample(5)
```

3. Decode listing category

```python
# Decode ListingCategory (numeric) feature
list_dict = {0 : 'Not Available', 1 : 'Debt Consolidation', 2 : 'Home Improvement', 3: 'Business', 
             4 : 'Personal Loan', 5 : 'Student Use', 6 : 'Auto', 7 : 'Other', 8 : 'Baby&Adoption', 
             9 : 'Boat', 10 : 'Cosmetic Procedure', 11 : 'Engagement Ring', 12 : 'Green Loans',
             13 : 'Household Expenses', 14 : 'Large Purchases', 15 : 'Medical/Dental', 16 : 'Motorcycle',
             17 : 'RV', 18 : 'Taxes', 19 : 'Vacation', 20 : 'Wedding Loans'}

dfClean['ListingCategory'] = dfClean['ListingCategory (numeric)'].map(list_dict)
```

## Multivariate Analysis


### Loan Outcome by various variables
1. By Credit Grade

```python
dfCreditGrade = dfClean.groupby(['ProsperScore','LoanStatus'])['LoanNumber'].count().unstack()
dfCreditGrade = dfCreditGrade.T.div(dfCreditGrade.T.sum()).T*100
dfCreditGrade.plot(kind='bar', stacked = True, title = 'Loan Status by Credit Grade');
print(dfClean['ProsperScore'].value_counts())
```

2. By Listing Category

```python
# This would remove categories with less than 100 completed loans
minObservations = 100
dfListingCategory = dfClean[dfClean['ListingCategory'].isin(dfClean['ListingCategory'].value_counts()[dfClean['ListingCategory'].value_counts() >=minObservations].index)]
dfListingCategory = dfListingCategory.groupby(['ListingCategory','LoanStatus'])['LoanNumber'].count().unstack()
dfListingCategory = dfListingCategory.T.div(dfListingCategory.T.sum()).T*100
dfListingCategory.plot(kind='bar', stacked = True, title = 'Loan Status by Listing Category');
print(dfClean.groupby('ListingCategory').filter(lambda x: len(x)>=minObservations)['ListingCategory'].value_counts())
```

3. Loan Amount

```python
dfLoanAmount = dfClean.groupby(['LoanOriginalAmountBin','LoanStatus'])['LoanNumber'].count().unstack()
dfLoanAmount = dfLoanAmount.T.div(dfLoanAmount.T.sum()).T*100
dfLoanAmount.plot(kind='bar', stacked = True, title = 'Loan Status by Loan Amount (Quartile)');
print(dfClean['LoanOriginalAmountBin'].value_counts())
```

4. Credit Lines

```python
dfCreditLines = dfClean.groupby(['CreditLinesBin','LoanStatus'])['LoanNumber'].count().unstack()
dfCreditLines = dfCreditLines.T.div(dfCreditLines.T.sum()).T*100
dfCreditLines.plot(kind='bar', stacked = True, title = 'Loan Status by # of Credit Lines');
print(dfClean['CreditLinesBin'].value_counts())
```

## Loan outcome by Prosper Score and Original Loan Amount
In this analysis, I want to see if there is a significant difference in loan size across prosper scores as I then want to see if higher or lower default rates are observed with large or small original loan amounts, after controlling for Prosper Score.

```python
#dfClean.groupby('ProsperScore')['LoanOriginalAmount'].describe()
base_color = sb.color_palette()[0]
sb.barplot(data = dfClean, x = 'ProsperScore', y='LoanOriginalAmount',color=base_color, ci = 'sd');
```

Based on the above plot, it looks like there is no meaniful difference in the mean loan size by Prosper Score. Given that the mean of each category is well within the standard deviation of the loan size for other Prosper Score categories (with the exception of Prosper Score == 11), this suggest that its okay to assume the difference in loan size by Prosper Score is statistically minor. As a result, we can move on with additional analysis.

```python
dfCreditScore_LoanSize = dfClean.groupby(['ProsperScore','LoanOriginalAmountBin','LoanStatus'])['LoanNumber'].count().unstack()
dfCreditScore_LoanSize = dfCreditScore_LoanSize.T.div(dfCreditScore_LoanSize.T.sum()).T*100


#https://stackoverflow.com/questions/54856726/pandas-how-to-group-by-and-plot-for-each-hour-of-each-day-of-week
for idx, gp in dfClean.groupby('ProsperScore'):    
    chartTitle = '(Prosper Score {0})'.format(idx)
    dfProsperScore = gp.groupby(['LoanOriginalAmountBin','LoanStatus'])['LoanNumber'].count().unstack()
    dfProsperScore = dfProsperScore.T.div(dfProsperScore.T.sum()).T*100
    dfProsperScore.plot(kind='bar', stacked = True, title = 'Loan Status by Original Loan Ammount {0}'.format(chartTitle));    
    
#dfCreditScore_LoanSize.plot(kind='bar', stacked = True, title = 'Loan Status by # of Credit Lines');
```
