```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from globalfunctions import *
%matplotlib inline

Ticker = 'SPY'
upperChannel = 252
lowerChannel = 252
percentThreshHold = 0.8
model = 'PercentFromHigh_' + str(percentThreshHold*100)
fileName = Ticker+'.csv'

df = pd.read_csv(fileName,sep=',')
df = df.dropna()

df.info()
```

```python
df.tail()
```

```python
def donchianChannel(upper,lower,df):
    highs = []
    lows = []
    i = 0
    nHigh = 0
    nLow = 0
    isWarm = False
    r = []
    
    for j,n in df.iterrows():
        highs.append(n['High'])
        lows.append(n['Low'])
        
        i +=1
        
        if i >= upper:
            highs.pop(0)
        if i >= lower:
            lows.pop(0)
        if i >= max(upper,lower) and isWarm==False:
            isWarm = True
            
        nHigh = max(highs)
        nLow = min(lows)
        close = n['Close']
        d = n['Date']
        if isWarm:
            r.append({'nHigh': nHigh,
                      'nLow':nLow,
                      'close':close,
                      'Date':d}
                    )
    print('donchianChannel Calculated')
    return r
        
        
```

```python
def percentFromHigh(lst,threshhold):
    positions = []
    for i in lst:
        positionCount = len(positions)
        if positionCount ==0:
            if i['close'] >= (threshhold*i['nHigh']):
                pos = Position(i['close'],i['Date'],0,'01/01/1900',0,'Long')
                positions.append(pos)
        else:
            latestPosition = positions[positionCount-1]
            if i['close'] >= (threshhold*i['nHigh']):
                if latestPosition.exitPrice >0:
                    pos = Position(i['close'],i['Date'],0,'01/01/1900',0,'Long')
                    positions.append(pos)
            else:
                if latestPosition.exitPrice == 0:
                    latestPosition.exitPrice = i['close']
                    latestPosition.exitDate = i['Date']
                    latestPosition.profitLoss = i['close']/latestPosition.entryPrice
                    positions.pop()
                    positions.append(latestPosition)
    return positions
                    
    

  
        
        
```

```python
obj = donchianChannel(upperChannel,lowerChannel,df)
positions = percentFromHigh(obj,percentThreshHold)
ModelAssessment(positions,Ticker,model)
```

```python
for p in positions:
    print(p.entryDate, p.entryPrice, p.exitDate,p.exitPrice, p.profitLoss)
```
